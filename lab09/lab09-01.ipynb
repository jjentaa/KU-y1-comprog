```python
import json

def read_json(filename):
    with open(filename) as f:
        data = f.read()
        data = json.loads(data)
    return data

filename = "IMDB_movies_merged.json"

data = read_json(filename)
```

```python
#Q1 What are the director names of the top five rating films of the actor Harrison Ford (who had been casted in 'Indiana Jones and the Raiders of the Lost Ark') that have not been directed by Steven Spielberg?
#Q2 What is the top rated film (not directed either by George Lucas or by Steven Spielberg) in which Harrison Ford and Tommy Lee Jones casting together?
```

```python
# sum all funtion
def q1():
    # find Harrison Ford and not Steven Spielberg as a director
    act_harr = []
    for i in data:
        try:
            cast_ls = i["cast"]
            director = i["director"]["name"]
            
            isHarrison = False
            for j in cast_ls:
                if(j["name"]=="Harrison Ford"):
                    isHarrison = True
                    break
            
            if(isHarrison and director!="Steven Spielberg"):
                act_harr.append(i)

        except:
            pass

    clean_ls = act_harr.copy()

    for ele in act_harr:
        if(ele["ratingValue"]==''): 
            clean_ls.remove(ele)

    all_rating = []

    for ele in clean_ls:
        try:
            num = float(ele["ratingValue"])
            all_rating.append(num)
        except:
            pass

    # # sorting
    # sorted_ls = sorted(clean_ls, key=lambda x: float(x['ratingValue'])) 
    # sorted_ls.reverse()

    all_rating = list(set(all_rating))

    all_rating.sort()
    all_rating.reverse()


    maxx = all_rating[0:5]

    maxx = [str(m) for m in maxx]


    ans = []
    for idx in clean_ls:
        if(idx["ratingValue"] in maxx):
            ans.append(idx["director"]["name"])

    ans.sort()

    for a in ans:
        print(a)
```

```python
q1()
```

```python
def q2():
    # find Harrison Ford and Tommy Lee Jones casting together and not directed either by George Lucas or by Steven Spielberg
    ls = []

    for i in data:
        try: 
            cast_ls = i["cast"]
            director = i["director"]["name"]
            isHarrison = False
            isTom = False
            for j in cast_ls:
                if(j['name']=="Harrison Ford"): isHarrison=True
                if(j['name']=="Tommy Lee Jones"): isTom=True

            if(isHarrison and isTom and (director not in ["George Lucas", "Steven Spielberg"])):
                ls.append(i)
        
        except:
            pass

    clean_ls = ls.copy()

    for ele in ls:
        if(i["ratingValue"]==''): 
            clean_ls.remove(ele)
    # # sorting
    sorted_ls = sorted(clean_ls, key=lambda x: float(x['ratingValue'])) 
    sorted_ls.reverse()

    print(sorted_ls[0]["name"])
```

```python
print("(1) Answer of Q1")
print("(2) Answer of Q2")

chioce = int(input("or just press (Enter): "))
if(chioce==1): q1()
if(chioce==2): q2()
```
